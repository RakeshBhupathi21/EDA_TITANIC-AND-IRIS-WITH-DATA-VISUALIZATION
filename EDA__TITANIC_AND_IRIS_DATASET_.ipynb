<h1>EDA OF IRIS DATASET WITH PYTHON.</h1>


<p>The Iris dataset was used in R.A Fisher's classic 1936 papper. The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower.One flower species is linearly separable from the other two, but the other two are not linearly separable from each other</p>

<h2>Import Libraries</h2>
<p>Let's import some libraries to get started!</p>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

<h2>The Data</h2>
<p>Let's start by reading in the Iris.csv file into pandas dataframe.</p>

```python
df = pd.read_csv("Iris.csv")
df
```

<h2>Exploratory Data Analysis</h2>
<p>Let's begin some exploratory data analysis! We'll start by checking the info of the data!</p>


```python
df.info()
```

<h2>Missing Data</h2>
<p>We can use seaborn to create a simple heatmap to see where we are missing data!</p>

```python
df.isnull()
```

```python
df.describe()
```

```python
df['Species'].value_counts()
```

```python
df.isnull().sum()
```

<h2>DATA VISUALISATION</h2>
<p>Let's continue on by visualizing some more of the data!</p>

```python
# Scatter plot
df.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue="Species",size=5) \
    .map(plt.scatter,"SepalLengthCm","SepalWidthCm") \
    .add_legend();
plt.show()
```

```python
del df['Id']
```

```python
# Pair plot
sns.set_style("whitegrid");
sns.pairplot(df,hue="Species",height=3);
plt.show()
```

```python
iris_setosa = df.loc[df["Species"]=="iris_setosa"];
iris_virginica = df.loc[df["Species"]=="iris_virginica"];
iris_versicolor = df.loc[df["Species"]=="iris_versicolor"];
```

```python
# Histogram
sns.FacetGrid(df,hue="Species",size=7) \
    .map(sns.distplot,"PetalLengthCm") \
    .add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue="Species",size=7) \
    .map(sns.distplot,"PetalWidthCm") \
    .add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue="Species",size=7) \
    .map(sns.distplot,"SepalLengthCm") \
    .add_legend();
plt.show()
```

```python
# Box plot
sns.boxplot(x='Species',y='PetalLengthCm',data=df)
plt.show()
```

```python
# Violen plot
sns.violinplot(x="Species",y="PetalLengthCm",data=df, height=9)
plt.show()
```
